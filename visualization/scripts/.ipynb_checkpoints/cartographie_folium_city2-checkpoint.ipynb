```python
#Chargement des librairies

import pandas as pd

import matplotlib.pyplot as plt  
from collections import Counter
import matplotlib.ticker as ticker

!pip install folium
import folium
```

```python
#Désignation du fichier de travail (tableur tsv)

path = '../data/PCP_complete.tsv'
```

```python
#Création d'un tableur (dataframe)

df = pd.read_csv(path, sep='\t', index_col=[0])
```

```python
#Aperçu des données 

df.head(3) 
```

```python
#Aperçu des données 

df.info()
```

```python
#Créer un nouveau tableur avec le compte du nombre d'items par ville des exposants

cities_nb = Counter(df["City of the artist"].sort_values(ascending=True).tolist()).most_common()
df_cities = pd.DataFrame(cities_nb, columns=["City of the artist", 'Count'])
df_cities
```

```python
df_cities = df_cities.set_index("City of the artist").join(df.set_index("City of the artist"), how="right").reset_index() 
df_cities.head()
```

```python
df_cities.drop_duplicates(subset=["City of the artist"], keep = 'first', inplace=True)
df_cities
```

```python
#Import de la librairie geopandas

import geopandas
```

```python
#Création des colones 'Latitude' et 'Longitude' à partir des colonnes 'City lat' et 'City long'

df_cities[['Latitude', 'Longitude']]= df_cities[['City lat', 'City long']]
df_cities.info(4)
```

```python
#Indiquer que notre tableur df_cities est un tableur géoréférencé 

gdf_cities = geopandas.GeoDataFrame(df_cities, geometry=geopandas.points_from_xy(df_cities.Longitude, df_cities.Latitude))
```

```python
# Définition du fond de carte

#Choix des coordonnées centrales
#loc = [27.783889, -97.510556]#centrement continent américain 
loc = [52.50, -13.35]#centrement Atlantique
title_html = '''
             <h3 align="center" style="font-size:16px"><b>Distribution of exhibitors' city according to the number of works exhibited
            </b></h3><h5 align="center">Frédérine Pradier - CC BY</h5> 
             '''.format(loc) +'\n'

animated_map = folium.Map(location=loc, tiles='Stamen toner', zoom_start=2)
animated_map.get_root().html.add_child(folium.Element(title_html))
animated_map
```

```python
gdf_cities = gdf_cities[gdf_cities["City of the artist"].notna()]
```

```python
def create_map_markers(gdf_cities, map_name):
    
    folium.CircleMarker(
        location=[gdf_cities['Latitude'], gdf_cities['Longitude']], 
        radius=0.03*gdf_cities['Count'],
        color="orange",
        fill=True,
        popup=folium.Popup(f"{gdf_cities['City of the artist'].title()} <br> {str(gdf_cities['Count']).title()}" +" items represented", max_width=200),
        tooltip=f"{gdf_cities['City of the artist'].title()}<br> {str(gdf_cities['Count']).title()}" +" items represented",
        data=gdf_cities,
        columns=['City of the artist', "Count"],
        legend_name="Number of items retrieved"
        
                 ).add_to(map_name)
```

```python
gdf_cities=gdf_cities.dropna(subset=['Latitude', 'Longitude'])
gdf_cities.apply(create_map_markers, map_name=animated_map, axis="columns")
animated_map.save('cities.html')
animated_map
```
